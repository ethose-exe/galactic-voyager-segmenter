# 🌌 Galactic Voyager Segmentation
### Objective: Identify high-impact passenger personas using K-Means Clustering.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('../data/voyagers.csv')
X = df[['Lightyears_Traveled', 'Luxury_Lounge_Spend']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = model.fit_predict(X_scaled)
print(df.head())
```
